```python
from transformers import AutoTokenizer
```

```python
model="distilbert-base-uncased-finetuned-sst-2-english"
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
tokenizer
```

```python
input=['Ea Nasir sold poor copper ingots','we were so lucky to find the complaint letter']
inputs=tokenizer(input,padding=True,truncation=True,return_tensors='pt')
inputs
```

```python
from transformers import AutoModel

model1='distilbert-base-uncased-finetuned-sst-2-english'
modele=AutoModel.from_pretrained(model1)
```

```python
outputs=modele(**inputs)
outputs.last_hidden_state.shape
```

```python
from transformers import AutoModelForSequenceClassification
modelss='distilbert-base-uncased-finetuned-sst-2-english'
model=AutoModelForSequenceClassification.from_pretrained(modelss)
outputs=model(**inputs)
```

```python
outputs['logits'].shape
```

```python
outputs.logits
```

```python
import torch
```

```python
preds=torch.nn.functional.softmax(outputs.logits,dim=-1)
preds
```

```python
model.config.id2label
```

```python
model.save_pretrained('directory')
```
