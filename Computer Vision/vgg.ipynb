```python
import torch
from torch import nn
```

```python
VGG_types={
    'VGG11':[64,'M',128,'M',256,256,'M',512,512,512,'M'],
    'VGG13':[64,64,'M',128,128,'M',256,256,'M',512,512,'M',512,512,'M'],
    'VGG16':[64,64,'M',128,128,'M',256,256,'M',512,512,512,'M',512,512,512,'M'],
    'VGG19':[64,64,'M',128,128,'M',256,256,256,'M',512,512,512,512,'M',512,512,512,512,'M']
}
```

```python
class VGG_net(nn.Module):
  def __init__(self,inn=3,numc=1000):
    super(VGG_net,self).__init__()
    self.inn=inn
    self.l1=self.create1(VGG_types['VGG16'])

    self.fcs=nn.Sequential(
        nn.Linear(512*7*7,4096),
        nn.ReLU(),
        nn.Dropout(p=0.5),
        nn.Linear(4096,4096),
        nn.ReLU(),
        nn.Dropout(p=0.5),
        nn.Linear(4096,numc)
    )
  def forward(self,x):
    x=self.l1(x)
    x=torch.flatten(x,1)
    x=self.fcs(x)

    return x

  def create1(self,vggtype):
    layer=[]
    inn=self.inn

    for x in vggtype:
      if type(x)==int:
        out=x
        layer+=[nn.Conv2d(in_channels=inn,out_channels=out,kernel_size=3,stride=1,padding=1),
                nn.BatchNorm2d(x),
                nn.ReLU()]
        inn=x

      elif x=='M':
        layer+=[nn.MaxPool2d(kernel_size=2,stride=2)]

    return nn.Sequential(*layer)
```
