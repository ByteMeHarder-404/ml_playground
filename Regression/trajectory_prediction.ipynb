```python
import os
import pandas as pd
```

```python
def datas(path):
  data=[]
  for file in os.listdir(path):
    df=pd.read_csv(os.path.join(path,file),sep=r'\s+',header=None)
    data.append(df)
  return pd.concat(data,ignore_index=True)
```

```python
train='/content/train'
test='/content/test'
```

```python
train=datas(train)
test=datas(test)
```

```python
train.head()
```

```python
train.columns=['Frame #','Aircraft ID','x (km)',	'y (km)',	'z (km)',	'windx (m/s)'	,'windy (m/s)',]
test.columns=['Frame #','Aircraft ID','x (km)',	'y (km)',	'z (km)',	'windx (m/s)'	,'windy (m/s)',]
```

```python
df=pd.concat([train,test])
```

```python
df.head()
```

```python
df=df.sort_values(['Aircraft ID','Frame #'])
```

```python
df['v_x']=df.groupby('Aircraft ID')['x (km)'].diff()
df['v_y']=df.groupby('Aircraft ID')['y (km)'].diff()
df['v_z']=df.groupby('Aircraft ID')['z (km)'].diff()
```

```python
df.head()
```

```python
df[['v_x','v_y','v_z']]=df[['v_x','v_y','v_z']].fillna(0)
```

```python
df['a_x']=df.groupby('Aircraft ID')['v_x'].diff()
df['a_y']=df.groupby('Aircraft ID')['v_y'].diff()
df['a_z']=df.groupby('Aircraft ID')['v_z'].diff()
df[['a_x','a_y','a_z']]=df[['a_x','a_y','a_z']].fillna(0)
```

```python
df.head()
```

```python
df['x_ideal'] = df['x (km)'] + df['v_x'] + 0.5*df['a_x']*1**2 + (df['windx (m/s)']/ 1000)
df['y_ideal'] = df['y (km)'] + df['v_y'] + 0.5*df['a_y']*1**2 + (df['windy (m/s)']/ 1000)
df['z_ideal'] = df['z (km)'] + df['v_z'] + 0.5*df['a_z']*1**2
```

```python
df.head()
```

```python
df['x_next'] = df.groupby('Aircraft ID')['x (km)'].shift(-1)
df['y_next'] = df.groupby('Aircraft ID')['y (km)'].shift(-1)
df['z_next'] = df.groupby('Aircraft ID')['z (km)'].shift(-1)
```

```python
df['res_x'] = df['x_next'] - df['x_ideal']
df['res_y'] = df['y_next'] - df['y_ideal']
df['res_z'] = df['z_next'] - df['z_ideal']
```

```python
df.head()
```

```python
df = df.dropna(subset=['res_x','res_y','res_z'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.ensemble import AdaBoostRegressor
import xgboost as xg
import lightgbm as lgbm
```

```python
features = ['x (km)', 'y (km)', 'z (km)','v_x', 'v_y', 'v_z','a_x', 'a_y', 'a_z','windx (m/s)', 'windy (m/s)']
```

```python
target_x = 'res_x'
target_y = 'res_y'
target_z = 'res_z'
```

```python
X_train, X_test, yx_train, yx_test = train_test_split(df[features], df[target_x], test_size=0.26, random_state=564)
_, _, yy_train, yy_test = train_test_split(df[features], df[target_y], test_size=0.26, random_state=564)
_, _, yz_train, yz_test = train_test_split(df[features], df[target_z], test_size=0.26, random_state=564)
```

```python
def train_and_evaluate(model, X_tr, X_te, y_tr, y_te, axis):
    model.fit(X_tr, y_tr)
    preds = model.predict(X_te)
    mse = mean_squared_error(y_te, preds)
    r2=r2_score(y_te,preds)
    print(f"{axis} | MSE: {mse:.2f} | R2:{r2}")
    return preds
```

```python
models = {
    "LGBM": lgbm.LGBMRegressor(n_estimators=200, learning_rate=0.05, max_depth=5),
    "XGB": xg.XGBRegressor(n_estimators=200, learning_rate=0.075, max_depth=6),
    "AB": AdaBoostRegressor(n_estimators=200, learning_rate=0.1)
}
```

```python
results={}
results['LGBM'] = train_and_evaluate(models['LGBM'], X_train, X_test, yx_train, yx_test, "X-axis")
results['XB'] = train_and_evaluate(models['XGB'], X_train, X_test, yy_train, yy_test, "Y-axis")
results['Ada'] = train_and_evaluate(models['AB'], X_train, X_test, yz_train, yz_test, "Z-axis")
```

```python
df_te=df.iloc[X_test.index].copy()
```

```python
df_te['x_preds']=df_te['x_ideal']+results['LGBM']
df_te['y_preds']=df_te['y_ideal']+results['XB']
df_te['z_preds']=df_te['z_ideal']+results['Ada']
df_te[['x (km)', 'y (km)', 'z (km)', 'x_preds', 'y_preds', 'z_preds']].head()
```
