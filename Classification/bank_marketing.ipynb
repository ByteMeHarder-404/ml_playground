```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/bank-full.csv',sep=';')
```

```python
df.head(10)
```

```python
for i in ['default','housing','loan']:
  df[i]=df[i].map({'no':0,'yes':1})
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
num = ['age','balance','duration','campaign','previous']
scaler = StandardScaler()
df[num] = scaler.fit_transform(df[num])
```

```python
df
```

```python
df['last_contact-days']=(df['pdays']!=-1).astype(int)
df['pdays_value'] = df['pdays'].apply(lambda x: 0 if x == -1 else x)
```

```python
df[['pdays_value']]=scaler.fit_transform(df[['pdays_value']])
```

```python
df['job'] = df['job'].astype('category').cat.codes
df['marital'] = df['marital'].astype('category').cat.codes
df['education'] = df['education'].astype('category').cat.codes
df['poutcome']=df['poutcome'].astype('category').cat.codes
```

```python
df
```

```python
df['y']=df['y'].map({'no':0,'yes':1})
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.25, random_state=746, stratify=df['y'])
```

```python
import torch
from torch import nn
abdcpp_train = torch.tensor(train_df[['age','balance','duration','campaign','previous','pdays_value']].values, dtype=torch.float32)
dhll_train = torch.tensor(train_df[['default','housing','loan','last_contact-days']].values, dtype=torch.float32)
jobs_train = torch.tensor(train_df['job'].values, dtype=torch.long)
shadi_train = torch.tensor(train_df['marital'].values, dtype=torch.long)
edu_train = torch.tensor(train_df['education'].values, dtype=torch.long)
out_train = torch.tensor(train_df['poutcome'].values, dtype=torch.long)
y_train = torch.tensor(train_df['y'].values, dtype=torch.float32)
```

```python
abdcpp_test = torch.tensor(test_df[['age','balance','duration','campaign','previous','pdays_value']].values, dtype=torch.float32)
dhll_test = torch.tensor(test_df[['default','housing','loan','last_contact-days']].values, dtype=torch.float32)
jobs_test = torch.tensor(test_df['job'].values, dtype=torch.long)
shadi_test = torch.tensor(test_df['marital'].values, dtype=torch.long)
edu_test = torch.tensor(test_df['education'].values, dtype=torch.long)
out_test = torch.tensor(test_df['poutcome'].values, dtype=torch.long)
y_test = torch.tensor(test_df['y'].values, dtype=torch.float32)
```

```python
class Model(nn.Module):
  def __init__(self,j,s,e,o):
    super().__init__()
    self.j_emb = nn.Embedding(j, 6)
    self.s_emb = nn.Embedding(s, 2)
    self.e_emb = nn.Embedding(e, 2)
    self.o_emb = nn.Embedding(o, 2)
    inn=6+2+2+2+6+4
    self.layer=nn.Sequential(
        nn.Linear(inn,32),
        nn.ReLU(),
        nn.Linear(32,64),
        nn.ReLU(),
        nn.Linear(64,8),
        nn.ReLU(),
        nn.Linear(8,1)
    )
  def forward(self,abdcpp,dhll,jobs,shadi,edu,out):
    j = self.j_emb(jobs)
    s = self.s_emb(shadi)
    e = self.e_emb(edu)
    o = self.o_emb(out)
    x = torch.cat([abdcpp, dhll, j, s, e, o], dim=1)
    return self.layer(x)
```

```python
model=Model(
    j=len(df['job'].unique()),
    s=len(df['marital'].unique()),
    e=len(df['education'].unique()),
    o=len(df['poutcome'].unique())
)
```

```python
model
```

```python
poss = y_train.sum()
neg = len(y_train) - poss
pos = torch.tensor([neg / poss], dtype=torch.float32)
lossf = nn.BCEWithLogitsLoss(pos_weight=pos)
opt = torch.optim.Adam(model.parameters(), 0.001)
sch = torch.optim.lr_scheduler.StepLR(opt, 100, 0.9)
```

```python
from sklearn.metrics import accuracy_score
torch.manual_seed(456)
lambda_l1 = 1e-5
epochs = 301

for epoch in range(epochs):
    model.train()
    preds = model(abdcpp_train, dhll_train, jobs_train, shadi_train, edu_train, out_train).squeeze()
    loss = lossf(preds, y_train)
    l1_penalty = sum(torch.sum(torch.abs(p)) for p in model.parameters())
    loss = loss + lambda_l1 * l1_penalty
    opt.zero_grad()
    loss.backward()
    opt.step()

    train_loss = loss.item()

    model.eval()
    with torch.no_grad():
        test_logits = model(abdcpp_test, dhll_test, jobs_test, shadi_test, edu_test, out_test).squeeze()
        test_probs = torch.sigmoid(test_logits)
        test_loss = lossf(test_logits, y_test).item()
        pred_labels = (test_probs >= 0.6).int().numpy()
        acc = accuracy_score(y_test.numpy(), pred_labels)

    sch.step()
    if epoch % 100 == 0:
        print(f"Epoch {epoch+1:02d}/{epochs} | "
              f"Train Loss: {train_loss:.4f} | "
              f"Test Loss: {test_loss:.4f} | "
              f"Test Acc: {acc:.4f} | "
              f"LR: {sch.get_last_lr()[0]:.5f}")
```

```python
from sklearn.metrics import classification_report
model.eval()
y_true, y_pred = [], []
with torch.no_grad():
    test_logits = model(abdcpp_test, dhll_test, jobs_test, shadi_test, edu_test, out_test).squeeze()
    test_loss = lossf(test_logits, y_test).item()
    test_probs = torch.sigmoid(test_logits)
    pred_labels = (test_probs >= 0.6).int().numpy()
    y_true = y_test.numpy()
    y_pred = pred_labels

print(f"Final Test Loss: {test_loss:.4f}")
print(classification_report(y_true, y_pred, target_names=["no", "yes"]))
```
